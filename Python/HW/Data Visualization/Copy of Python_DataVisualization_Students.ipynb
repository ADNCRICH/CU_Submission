```python
# Create index of 100 students and save it in a list 
index_student = [idx+1 for idx in range(len(gender))]
#print(index_student) #นักเรียน 100 คน

#ดึงค่าคะแนนของนักเรียนทั้ง 3 วิชามาจากDataFrameของPandas
math_score = csv_data['math score']
reading_score = csv_data['reading score']
writing_score = csv_data['writing score']
```

```python
#เรียงค่าคะแนนวิชาคณิตศาตร์ และวิชาการอ่านจากน้อยไปหามาก
#วิชาคณิตศาสตร์ 
#Write code below please.
math_score = sorted(math_score.items(), key=lambda item:item[1])
print(math_score[:10])
#วิชาการอ่าน
#Write code below please.
reading_score = sorted(reading_score.items(),key=lambda item:item[1])
print(reading_score[:10])
writing_score = sorted(writing_score.items(),key=lambda item:item[1])
```

```python
plt.figure(figsize=(10,6))
# Write code below please. #
MS=math_score[:10]
MS=[i for j,i in MS]
RS=reading_score[:10]
RS=[i for j,i in RS]
idx=[i for i in range(10)]
plt.plot(idx,MS,color='mediumaquamarine')
plt.plot(idx,RS,color='mediumslateblue')
############################
plt.title('Line chart comparison on Math and Reading score of 10 students')
plt.ylabel('Score')
plt.xlabel('Individual student')
```

# TO DO 5: หาค่ามัธยฐานและค่าเฉลี่ยของแต่ละวิชา และนำไปแสดงผลบนกราฟแท่งแบบ Multiple plot 

<p> ภาพตัวอย่าง</p>
<img src="https://github.com/thcktw/python-com-resource/raw/master/lab4datavis/result_of_todo5_new.png">

```python
#Write code below please.
mean_math = sum(i[1] for i in math_score)/len(math_score)
mean_writing = sum(i[1] for i in writing_score)/len(writing_score)
mean_reading = sum(i[1] for i in reading_score)/len(reading_score)
median_math = (math_score[(len(math_score)-1)//2][1]+math_score[(len(math_score))//2][1])/2
median_reading = (reading_score[(len(reading_score)-1)//2][1]+reading_score[(len(reading_score))//2][1])/2
median_writting = (writing_score[(len(writing_score)-1)//2][1]+writing_score[(len(writing_score))//2][1])/2
Mean = [mean_math,mean_writing,mean_reading]
Median = [median_math,median_reading,median_writting]
```

```python
#Plot multiple bar graph
#Write code below please.
plt.figure(figsize=(13,8))

idx_mean=[i for i in range(0,3)]
idx_median=[i+0.4 for i in range(0,3)]
idx_mid=[i+0.2 for i in range(0,3)]
Name = ['Math','Reading','Writing']

plt.bar(idx_mean,Mean,0.4,color='springgreen',label='Mean score')
plt.bar(idx_median,Median,0.4,color='khaki',label='Median score')

plt.title('Scores by group')
plt.ylabel('Score')
plt.xlabel('Subject')
plt.xticks(idx_mid,Name)
plt.legend(loc='upper left')
plt.show()
```
